```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.pyplot as plt
import numpy as np

# Define x-values corresponding to the three states
x_points = np.array([0, 0.5, 1.0])

# Data for each decision
decision_A = np.array([3.95, 3.40, 3.20])
decision_B = np.array([3.82, 3.80, 3.68])

# Calculate the intersection point on the first segment (between x=0 and x=0.5)
slope_A = (decision_A[1] - decision_A[0]) / (x_points[1] - x_points[0])
slope_B = (decision_B[1] - decision_B[0]) / (x_points[1] - x_points[0])
x_intersect = (decision_A[0] - decision_B[0]) / (slope_B - slope_A)
y_intersect = decision_A[0] + slope_A * x_intersect

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_points, decision_A, marker='o', linestyle='-', label='Option 1 - Contingency Financing')
plt.plot(x_points, decision_B, marker='s', linestyle='-', label='Option 2 - Improving Up-Front Planning')

# Mark and annotate the intersection point
plt.scatter(x_intersect, y_intersect, color='red', zorder=5)
plt.text(x_intersect, y_intersect, f'  p={x_intersect:.2f}', color='red', fontsize=10)

# Set x-axis range from 0 to 1 with increments of 0.2
plt.xticks(np.arange(0, 1.01, 0.2))
plt.xlim(0, 1)

# Adding labels and title with updated axis and graph titles
plt.xlabel('Probability (p)')
plt.ylabel('Expected Value (EV)')
plt.title('Sensitivity Analysis for Project Cancellations')
plt.grid(True)
plt.legend()

plt.show()



# Define x-values corresponding to the three states
x_points = np.array([0, 0.5, 1.0])

# Data for each decision
decision_A = np.array([3.80, 3.60, 3.50])
decision_B = np.array([3.88, 3.68, 3.40])

# Calculate slopes for the second segment (between x=0.5 and x=1.0)
slope_A2 = (decision_A[2] - decision_A[1]) / (x_points[2] - x_points[1])
slope_B2 = (decision_B[2] - decision_B[1]) / (x_points[2] - x_points[1])
denom2 = slope_A2 - slope_B2

# Compute the intersection point for the second segment if slopes differ
if np.isclose(denom2, 0):
    print("The lines are parallel or nearly parallel on the second segment; no unique intersection exists.")
    x_intersect = None
    y_intersect = None
else:
    # Calculate the intersection x-value relative to x=0.5
    x_intersect = (decision_B[1] - decision_A[1]) / denom2 + x_points[1]
    y_intersect = decision_A[1] + slope_A2 * (x_intersect - x_points[1])

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_points, decision_A, marker='o', linestyle='-', label='Decision 1 - Inventory & Supplier Diversification')
plt.plot(x_points, decision_B, marker='s', linestyle='-', label='Decision 2 - Training & Development of Workers')

# Mark and annotate the intersection point if valid and within x=0.5 to 1
if x_intersect is not None and x_points[1] <= x_intersect <= x_points[2]:
    plt.scatter(x_intersect, y_intersect, color='red', zorder=5)
    plt.text(x_intersect, y_intersect, f'  p={x_intersect:.2f}', color='red', fontsize=10)

# Set x-axis range from 0 to 1 with increments of 0.2
plt.xticks(np.arange(0, 1.01, 0.2))
plt.xlim(0, 1)

# Adding labels and title with updated axis and graph titles
plt.xlabel('Probability (p)')
plt.ylabel('Expected Value (EV)')
plt.title('Sensitivity Analysis - Project Delays')
plt.grid(True)
plt.legend()

plt.show()
```
